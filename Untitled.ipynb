```python
from collections import Counter
```

```python
vocab = Counter()
import re
```

```python
f = open('train.txt','r')
```

```python
sentences = []
sentence = []
ner_set = set()
```

```python
for line in f :
    if len(line) == 0 or line.startswith('-DOCSTART') or line[0] == '\n':
        if len(sentence) > 0 :
            sentences.append(sentence)
            sentence= []
        continue
    splits = line.split(' ')
    splits[-1] = re.sub(r'\n','',splits[-1])
    word = splits[0].lower()
    vocab[word] = vocab[word]+1
    sentence.append([word,splits[-1]])
    ner_set.add(splits[-1])
sentences[:3]
```

```python
print(ner_set)
```

```python
vocab_sorted = sorted(vocab.items(), key = lambda x:x[1], reverse = True)
vocab_sorted
```

```python
word_to_index = {w: i+2 for i, (w,n) in enumerate(vocab_sorted) if n >5}
word_to_index['PAD'] = 0
word_to_index['OOV']=1
word_to_index
```

```python
ner_to_index = {}
ner_to_index['PAD'] = 0
i=1
for ner in ner_set :
    ner_to_index[ner] = i
    i = i+1
print(ner_to_index)
```

```python
data_X = []

for s in sentences :
    temp_X = []
    for w, label in s :
        try : 
            temp_X.append(word_to_index.get(w,1))
        except KeyError:
            temp_X.append(word_to_index['OOV'])
    data_X.append(temp_X)
print(data_X)
```
